```python
import numpy as np
import matplotlib.pyplot as plt

# Vehicle Weight in kg (X)
x = np.array([105, 174, 1025, 1252, 1835, 2163, 2711])

# Fuel consumption L/100km (Y)
y = np.array([1.6, 3.6, 5.4, 7, 9.7, 10.5, 14.4])

# Perform linear regression
coefs = np.polyfit(x, y, 1)
regression_line = np.poly1d(coefs)

# Generate x values for the regression line
x_reg = np.linspace(min(x), max(x), 100)

# Calculate the corresponding y values for the regression line
y_reg = regression_line(x_reg)

# Plot the scatter plot
plt.scatter(x, y)

# Plot the regression line
plt.plot(x_reg, y_reg, color='red')

# Adding title to the plot
plt.title("Vehicle Weight vs Fuel Consumption")

# Adding label on the y-axis
plt.ylabel('Fuel Consumption (L/100km)')

# Adding label on the x-axis
plt.xlabel('Vehicle Weight (kg)')

plt.show()
```
