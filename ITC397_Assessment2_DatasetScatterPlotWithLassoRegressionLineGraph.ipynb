```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso

# Vehicle Weight in kg (X)
x = np.array([105, 174, 1025, 1252, 1835, 2163, 2711])

# Fuel consumption L/100km (Y)
y = np.array([1.6, 3.6, 5.4, 7, 9.7, 10.5, 14.4])

# Reshape X and Y arrays to be 2D
X = np.array(x).reshape(-1, 1)
Y = np.array(y).reshape(-1, 1)

# Create a Lasso model with a specified alpha value
alpha = 1
lasso = Lasso(alpha=alpha, fit_intercept=True)

# Fit the Lasso model to the data
lasso.fit(X, Y)

# Generate X values for the Lasso regression line
X_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)

# Calculate corresponding Y values using the Lasso model
Y_line = lasso.predict(X_line)

# Plot the scatter plot
plt.scatter(x, y)

# Plot the regression line
plt.plot(X_line, Y_line, color='red', label=f'Lasso Regression (alpha={alpha})')

# Adding title to the plot
plt.title("Vehicle Weight vs Fuel Consumption")

# Adding label on the y-axis
plt.ylabel('Fuel Consumption (L/100km)')

# Adding label on the x-axis
plt.xlabel('Vehicle Weight (kg)')

# Adding legend
plt.legend()

plt.show()
```
